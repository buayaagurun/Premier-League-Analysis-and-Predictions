```python
match_new = "/content/drive/MyDrive/LATIHAN PYHTON/output_new.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(match, index_col=False)
df2 = pd.read_csv(match_new, index_col=False)
```

```python
df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed')]
df2 = df2.loc[:, ~df2.columns.str.contains('^Unnamed')]
```

```python
df2 = df2.replace(2022,2023)
```

```python
df = df1.append(df2, ignore_index=True)
```

```python
df
```

```python
df.columns
```

1. THE MOST DANGEROUS TEAM ON ATTACK

The criteria is, how many goal, shoot, shoot on target

2. THE MOST EFFECTIVE TEAM ON ATTACK

The criteria is how goal/xg, goal/shoot, goal/shoot on target

##2021/2022

```python
result_2022 = df[df['season']==2022]
```

```python
result_2022
```

```python
stats_2022 = result_2022.groupby('team', as_index=False).sum()
```

```python
stats_2022["xg/gf"] = stats_2022["xg"]/stats_2022["gf"]
stats_2022["sh/gf"] = stats_2022["sh"]/stats_2022["gf"]
stats_2022["sot/gf"] = stats_2022["sot"]/stats_2022["gf"]
```

```python
stats_2022
```

GRAPHIC 1, The most GOAL, SHOOT, SHOOT ON TARGET

```python
#GOAL
goal = stats_2022.sort_values("gf", ascending=True)
shoot = stats_2022.sort_values("sh", ascending=True)
sot = stats_2022.sort_values("sot", ascending=True)
xg = stats_2022.sort_values("xg", ascending=True)
```

```python
plt.figure(figsize=(12,5))
plt.barh("team","gf", data=goal)
plt.xlabel('GOAL', size=14)
plt.xticks(size=12)

plt.figure(figsize=(12,5))
plt.barh("team","sh", data=shoot)
plt.xlabel('SHOOT', size=14)
plt.xticks(size=12)

plt.figure(figsize=(12,5))
plt.barh("team","sot", data=sot)
plt.xlabel('SHOOT ON TARGET', size=14)
plt.xticks(size=12)

plt.figure(figsize=(12,5))
plt.barh("team","xg", data=xg)
plt.xlabel('EXPECTED GOAL', size=14)
plt.xticks(size=12)


plt.show()
```

So, Manchester City and Liverpool are the top 2 teams that have the most end product in Premier League last season. We can look on the graph above, that ManCity and Liverpool are so dominant they have almost 100 goals, +- 700 shoot and >200 shoot on target, that mean both of that team are so offensive and dangerous team on attacking side

```python
#GOAL
xg_gf = stats_2022.sort_values("xg/gf", ascending=True)
sh_gf = stats_2022.sort_values("sh/gf", ascending=True)
sot_gf = stats_2022.sort_values("sot/gf", ascending=True)
```

```python
plt.figure(figsize=(12,5))
plt.barh("team","sh/gf", data=sh_gf)
plt.xlabel('GOAL/SHOOT', size=14)
plt.xticks(size=12)

plt.figure(figsize=(12,5))
plt.barh("team","sot/gf", data=sot_gf)
plt.xlabel('GOAL/SHOOT ON TARGET', size=14)
plt.xticks(size=12)

plt.figure(figsize=(12,5))
plt.barh("team","xg/gf", data=xg_gf)
plt.xlabel('GOAL/EXPECTED GOAL', size=14)
plt.xticks(size=12)

plt.show()
```
